```python
import mglearn
```

```python
# 1. 차원 축소 (Dimensionality Reduction)
# 주성분 분석(PCA)이나 비음수 행렬 분해(NMF)와 같은 차원 축소 알고리즘의 원리를 보여줍니다.
```

```python
# 용도: PCA가 데이터의 분산이 가장 큰 방향(주성분)을 찾고, 데이터를 회전시킨 후 정보를 추출하는 과정을 단계별로 보여줍니다.
mglearn.plots.plot_pca_illustration()
```

```python
# 용도: NMF가 양수 성분을 바탕으로 데이터를 어떻게 재구성하는지 시각화합니다.
mglearn.plots.plot_nmf_illustration()
```

```python
# 용도: PCA 화이트닝(Whitening) 과정을 통해 데이터의 상관관계를 제거하고 분산을 균일하게 만드는 효과를 보여줍니다.
mglearn.plots.plot_pca_whitening()
```

```python
# 2. 군집화 (Clustering)
# 군집화 알고리즘이 클러스터를 형성해 나가는 반복적인 과정을 시각적으로 추적할 수 있습니다.
# K-평균 (K-Means)
```

```python
# 용도: 초기 중심점 설정 -> 할당 -> 중심점 재계산으로 이어지는 K-Means의 반복 수행 과정을 한눈에 보여줍니다.
mglearn.plots.plot_kmeans_algorithm()
```

```python
# 용도: K-Means가 학습한 결과로 만들어진 **클러스터 경계면(보로노이 다이어그램)**을 시각화합니다.
mglearn.plots.plot_kmeans_boundaries()
```

```python
# 계층적 군집 (Agglomerative Clustering)
# 용도: 가장 가까운 포인트들이 하나씩 묶여가며 거대한 군집을 형성하는 상향식(Bottom-up) 과정을 보여줍니다.
mglearn.plots.plot_agglomerative_algorithm()
```

```python
# 용도: 최종 군집 결과를 트리 구조와 유사하게 시각화합니다.
mglearn.plots.plot_agglomerative()
```

```python
# DBSCAN
# 용도: 매개변수 eps와 min_samples의 변화에 따라 핵심 샘플(Core points), 경계 샘플, **잡음(Noise)**이 어떻게 달라지는지 비교하여 보여줍니다.
mglearn.plots.plot_dbscan()
```

```python
# 설명: 비지도 학습 전 필수 단계인 **데이터 스케일링(StandardScaler, MinMaxScaler 등)**이 데이터 분포를 어떻게 바꾸는지 비교해 줍니다.
mglearn.plots.plot_scaling()
```
