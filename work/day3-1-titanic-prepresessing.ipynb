```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df.info()
# sex(Dtype이 object)을 숫자로 바꾸자.
# age의 263개의 결측치를 채우자
# fare(운임)의 1개의 결측치는 버리자.
```

```python
# sex을 숫자로 변경
# df['추가할_컬럼'] = 계산식
df['sex2'] = df['sex'].map( {'female':1, 'male':0} )
df
```

```python
# 운임(fare)의 결측치 삭제
df = df.dropna(subset=['fare'])
df
```

```python
# name 컬럼에서 호칭을 뽑아내자. 예) Miss. ==> Miss
# df['추가할_컬럼'] = 계산식
df['title'] = df['name'].str.extract("([A-Za-z]+)\.") # 정규표현식
df
```

```python
# 5개로 통합 : Master(아이), Miss, Mr, Mrs, Other(기타)
before = ['Capt', 'Col', 'Countess', 'Don', 'Dona',
          'Dr', 'Jonkheer', 'Lady', 'Major', 'Mlle',
          'Mme', 'Ms', 'Rev', 'Sir']
after  = ['Mr', 'Mr', 'Mrs', 'Mr', 'Miss',
          'Other', 'Mr', 'Mrs', 'Mr', 'Miss',
          'Mrs', 'Miss', 'Other', 'Mr']
df['title2'] = df['title'].replace( before, after )
df
```

```python
df['title2'].value_counts()
```

```python
# 예) title2 가 'Master' 이면서(&) 나이(age)가 결측치인 데이터 확인
# df.loc[ 조건, 컬럼명 ]
df.loc[ (df['title2'] == 'Master') & (df['age'].isnull()),  ]
```

```python
df.loc[ (df['title2'] == 'Master') & (df['age'].isnull()), 'age' ]
```

```python
# 호칭(title2)별 나이의 평균
df.groupby('title2')['age'].mean()
```

```python
df.loc[ (df['title2'] == 'Master') & (df['age'].isnull()), 'age' ] = 5.48
df.loc[ (df['title2'] == 'Miss') & (df['age'].isnull()), 'age' ] = 21.9
df.loc[ (df['title2'] == 'Mr') & (df['age'].isnull()), 'age' ] = 32.52
df.loc[ (df['title2'] == 'Mrs') & (df['age'].isnull()), 'age' ] = 36.95
df.loc[ (df['title2'] == 'Other') & (df['age'].isnull()), 'age' ] = 42.33
```

```python
df.isnull().sum() # age    0 <== age 결측치 채우기 성공
```

```python
# 저장
df.to_csv('titanic_ok.csv', index=False)
```
