1. 결과 및 식별 정보
- PassOrFail: 제품의 합격(0) 또는 불합격(1) 여부를 나타내는 타겟 변수입니다. (데이터 상 불합격률은 약 0.58%로 매우 낮습니다.)

2. 공정 시간 (Time 관련)
- Injection_Time: 용융된 플라스틱을 금형 안으로 밀어 넣는 데 걸린 시간입니다.
- Filling_Time: 금형 내부의 캐비티(빈 공간)가 완전히 채워지는 데 걸린 시간입니다.
- Plasticizing_Time: 다음 사출을 위해 플라스틱 원료를 녹여 준비하는(가소화) 시간입니다.
- Cycle_Time: 사출, 냉각, 제품 취출 등 전체 한 사이클이 완료되는 데 걸린 총 시간입니다.
- Clamp_Close_Time: 금형이 닫히는 데 소요된 시간입니다.

3. 위치 및 거리 (Position 관련)
- Cushion_Position: 사출 및 보압 완료 후 스크류 끝에 남은 수지의 위치입니다. 제품의 일정함을 판단하는 중요한 지표입니다.
- Plasticizing_Position: 가소화 과정이 끝났을 때 스크류의 위치입니다.
- Clamp_Open_Position: 제품을 꺼내기 위해 금형이 열렸을 때의 위치입니다.

4. 속도, 압력 및 RPM (Speed, Pressure, RPM 관련)
- Max_Injection_Speed: 사출 과정 중 스크류의 최대 이동 속도입니다.
- Max_Screw_RPM / Average_Screw_RPM: 플라스틱을 녹일 때 스크류가 회전하는 속도(분당 회전수)의 최대값과 평균값입니다.
- Max_Injection_Pressure: 사출 시 가해진 최대 압력입니다.
- Max_Switch_Over_Pressure: 충전 단계에서 보압 단계로 전환되는 시점의 압력입니다.
- Max_Back_Pressure / Average_Back_Pressure: 가소화 시 스크류가 뒤로 밀려날 때 가해지는 저항 압력(배압)의 최대값과 평균값입니다.

5. 온도 (Temperature 관련)
- Barrel_Temperature_1 ~ 6: 플라스틱이 녹는 통(배럴)의 구역별 온도입니다. (1번부터 6번까지 각 구간의 가열 상태를 나타냅니다.)
- Hopper_Temperature: 원료가 투입되는 입구(호퍼) 부근의 온도입니다.
- Mold_Temperature_3 ~ 4: 금형 내부 특정 지점의 온도입니다. 제품의 냉각 속도에 영향을 미칩니다.

```python
# ---------------------------------------------------------
# 14. 상관분석
# ---------------------------------------------------------
corr = df[numeric_cols].corr()

plt.figure(figsize=(20, 12))
sns.heatmap(corr, cmap="coolwarm", center=0, linewidths=0.1, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# ---------------------------------------------------------
# 15. 상관관계가 높은 변수쌍 상위 15개 추출
# ---------------------------------------------------------

# 1. 상관계수 행렬 계산 및 절대값 변환 (음의 상관관계도 중요하니까요!)
corr_matrix = df.corr().abs()

# 2. 행렬 형태를 리스트 형태로 풀기 (Stack)
# 예: (A, B) = 0.8 형태로 변환
pairs = corr_matrix.unstack().reset_index() # 인덱스(Index)를 컬럼(Columns) 방향으로 펼침
pairs.columns = ["변수1", "변수2", "상관계수"]

# 3. 중복 제거 및 자기 자신 제거 (핵심 포인트!)
# '변수1 < 변수2' 조건을 주면 (A, B)는 남고 (B, A)와 (A, A)는 자동으로 필터링됩니다.
# 대학생 수준에서 가장 이해하기 쉽고 깔끔한 트릭입니다.
top_pairs = pairs[pairs["변수1"] < pairs["변수2"]].copy()

# 4. 상관계수 기준으로 내림차순 정렬
top_pairs = top_pairs.sort_values(by="상관계수", ascending=False)

# 5. 결과 출력
print("🔍 상관계수 상위 15개 변수쌍:")
display(top_pairs.head(15).reset_index(drop=True))
```

```python
# ---------------------------------------------------------
# 16. 정답(PassOrFail)에 영향을 많이 주는 변수 찾기
# ---------------------------------------------------------

target_col = 'PassOrFail'

# 1. 전체 상관계수 표에서 'PassOrFail' 열만 따로 떼어냅니다.
target_corr = corr[target_col]

# 2. 자기 자신('PassOrFail' vs 'PassOrFail')은 항상 1이므로 분석에서 제외합니다.
target_corr = target_corr.drop(target_col)

# 3. 상관관계의 '세기'를 기준으로 정렬합니다.
# [중요] 상관계수는 -0.9처럼 음수여도 관계가 매우 강한 것입니다.
# 따라서 '절대값(abs)'을 기준으로 내림차순 정렬하여 가장 영향력 있는 변수를 맨 위로 올립니다.
sorted_indices = target_corr.abs().sort_values(ascending=False).index
final_result = target_corr.loc[sorted_indices]

# 4. 보기 좋게 표(DataFrame) 형태로 출력합니다.
print(f"🔍 {target_col}와의 상관계수 (영향력 순서)")
display(final_result.to_frame(name="상관계수").head(10))
```

```python
plt.figure(figsize=(10, 8))
ax = target_corr.sort_values().plot(kind="barh")
for i in ax.containers:
    ax.bar_label(i, fmt="{:.2f}", label_type='edge', padding=1, fontsize=8, color="black", fontweight="bold")
plt.xlim( -0.4, 0.5 )
plt.title("PassOrFail와의 상관계수")
plt.xlabel("상관계수")
plt.show()
```

```python
# ---------------------------------------------------------
# 16. PassOrFail 그룹별 상세 통계 비교
# ---------------------------------------------------------
print("=== PassOrFail 그룹별 기술통계량 ===")
group_desc = df.groupby(target_col)[feature_cols].describe()
display(group_desc)
```

```python
# ---------------------------------------------------------
# 17. 간단한 이상치 확인(IQR 기준) - 변수별 이상치 개수
# ---------------------------------------------------------
outlier_summary = []

for col in feature_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()

    outlier_summary.append({
        "variable": col,
        "outlier_count": outlier_count,
        "outlier_ratio_percent": round(outlier_count / len(df) * 100, 3)
    })

outlier_df = pd.DataFrame(outlier_summary).sort_values("outlier_count", ascending=False)

print("=== 변수별 이상치 개수(IQR 기준) ===")
display(outlier_df.head(15))
```

```python
# ---------------------------------------------------------
# 18. 정상/불량 평균 차이 해석용 표 정리
# ---------------------------------------------------------
eda_summary = group_mean.sort_values("abs_diff", ascending=False).copy()
eda_summary["mean_pass_0"] = eda_summary["mean_pass_0"].round(3)
eda_summary["mean_fail_1"] = eda_summary["mean_fail_1"].round(3)
eda_summary["diff_fail_minus_pass"] = eda_summary["diff_fail_minus_pass"].round(3)
eda_summary["abs_diff"] = eda_summary["abs_diff"].round(3)

print("=== EDA 핵심 요약표 ===")
display(eda_summary.head(15))
```

```python
# ---------------------------------------------------------
# 19. 해석용 문장 자동 출력
# ---------------------------------------------------------
print("=== 평균 차이가 큰 상위 5개 변수 해석 ===")
for col in eda_summary.head(5).index:
    pass_mean = eda_summary.loc[col, "mean_pass_0"]
    fail_mean = eda_summary.loc[col, "mean_fail_1"]

    if fail_mean > pass_mean:
        print(f"- {col}: 불량(1) 평균이 정상(0)보다 높습니다. ({fail_mean} > {pass_mean})")
    else:
        print(f"- {col}: 불량(1) 평균이 정상(0)보다 낮습니다. ({fail_mean} < {pass_mean})")
```
