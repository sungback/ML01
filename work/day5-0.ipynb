```python
# =========================================================
# Day 5. 시계열 분석과 예지보전 실습
# Python + pandas + statsmodels + scikit-learn
# Jupyter에 한 번에 붙여넣고 실행 가능
#
# 구성
# 1) 가상 제조 시계열 데이터 생성
# 2) 시계열 구조 확인
# 3) 추세 / 계절성 / 분해
# 4) 정상성 확인(ADF test)
# 5) ARIMA / SARIMA
# 6) ML 기반 시계열 예측 (lag, rolling feature)
# 7) 간단한 이상징후 탐색
# 8) 성능 비교 및 해석
# =========================================================

# ---------------------------------------------------------
# 1. 라이브러리 불러오기
# ---------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ---------------------------------------------------------
# 2. 기본 설정
# ---------------------------------------------------------
plt.rcParams["figure.figsize"] = (10, 5)
np.random.seed(42)

# ---------------------------------------------------------
# 3. 가상 제조 시계열 데이터 생성
#    - 실제 제조 진동/온도/설비상태 데이터가 아직 확정되지 않았을 때
#      수업용으로 가장 안정적인 방식
# ---------------------------------------------------------
date_index = pd.date_range(start="2024-01-01", periods=365, freq="D")

trend = np.linspace(50, 70, 365)                         # 장기 추세
seasonality = 8 * np.sin(2 * np.pi * np.arange(365) / 30)  # 월간 주기 비슷한 반복
noise = np.random.normal(0, 2.5, 365)                   # 잡음

# 예측 대상: 설비 진동값 또는 생산 KPI라고 가정
target_value = trend + seasonality + noise

# 일부 이상 구간(설비 이상징후라고 가정) 추가
target_value[120:126] += 12
target_value[250:255] += 15

df = pd.DataFrame({
    "timestamp": date_index,
    "target_value": target_value
})

# 추가 센서 변수 생성
df["temperature"] = 60 + 0.2 * np.arange(365) / 10 + 3 * np.sin(2 * np.pi * np.arange(365) / 14) + np.random.normal(0, 1.2, 365)
df["pressure"] = 30 + 2 * np.sin(2 * np.pi * np.arange(365) / 7) + np.random.normal(0, 0.8, 365)
df["rpm"] = 1000 + 20 * np.sin(2 * np.pi * np.arange(365) / 10) + np.random.normal(0, 10, 365)

print("=== 데이터 상위 5행 ===")
display(df.head())

print("\n=== 데이터 크기 ===")
print(df.shape)

print("\n=== 데이터 정보 ===")
df.info()

# ---------------------------------------------------------
# 4. timestamp 처리
# ---------------------------------------------------------
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp").reset_index(drop=True)

print("\n=== timestamp 범위 ===")
print(df["timestamp"].min(), "to", df["timestamp"].max())

# ---------------------------------------------------------
# 5. 기본 시계열 시각화
# ---------------------------------------------------------
plt.figure(figsize=(14, 5))
plt.plot(df["timestamp"], df["target_value"])
plt.title("Target Time Series")
plt.xlabel("Date")
plt.ylabel("Target Value")
plt.show()

plt.figure(figsize=(14, 4))
plt.plot(df["timestamp"], df["temperature"], label="temperature")
plt.plot(df["timestamp"], df["pressure"], label="pressure")
plt.plot(df["timestamp"], df["rpm"], label="rpm")
plt.title("Additional Sensor Variables")
plt.xlabel("Date")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 6. 기초 통계 확인
# ---------------------------------------------------------
print("\n=== 기초 통계량 ===")
display(df[["target_value", "temperature", "pressure", "rpm"]].describe().T)

# ---------------------------------------------------------
# 7. 이동평균 / 이동표준편차
#    - 최근 구간의 흐름과 변동성 보기
# ---------------------------------------------------------
df["rolling_mean_7"] = df["target_value"].rolling(window=7).mean()
df["rolling_std_7"] = df["target_value"].rolling(window=7).std()

plt.figure(figsize=(14, 5))
plt.plot(df["timestamp"], df["target_value"], label="target_value", alpha=0.6)
plt.plot(df["timestamp"], df["rolling_mean_7"], label="rolling_mean_7", linewidth=2)
plt.title("Target with 7-day Rolling Mean")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()

plt.figure(figsize=(14, 5))
plt.plot(df["timestamp"], df["rolling_std_7"], label="rolling_std_7")
plt.title("7-day Rolling Standard Deviation")
plt.xlabel("Date")
plt.ylabel("Std")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 8. 시계열 분해
#    - 추세 / 계절성 / 잔차 확인
# ---------------------------------------------------------
decompose_result = seasonal_decompose(df["target_value"], model="additive", period=30)
decompose_result.plot()
plt.suptitle("Seasonal Decompose (period=30)", y=1.02)
plt.show()

# ---------------------------------------------------------
# 9. 정상성 확인 - ADF test
# ---------------------------------------------------------
adf_result = adfuller(df["target_value"])

print("\n=== ADF Test (원시 시계열) ===")
print("ADF Statistic :", round(adf_result[0], 4))
print("p-value       :", round(adf_result[1], 4))
print("해석: p-value가 0.05보다 작으면 정상성 있다고 볼 가능성이 큼")

# 1차 차분
df["target_diff_1"] = df["target_value"].diff()

adf_result_diff = adfuller(df["target_diff_1"].dropna())

print("\n=== ADF Test (1차 차분 후) ===")
print("ADF Statistic :", round(adf_result_diff[0], 4))
print("p-value       :", round(adf_result_diff[1], 4))

plt.figure(figsize=(14, 5))
plt.plot(df["timestamp"], df["target_diff_1"])
plt.title("1st Differenced Series")
plt.xlabel("Date")
plt.ylabel("Differenced Value")
plt.show()

# ---------------------------------------------------------
# 10. train / test 분리
#    - 시계열은 시간 순서 유지가 핵심
# ---------------------------------------------------------
split_idx = int(len(df) * 0.8)

train_df = df.iloc[:split_idx].copy()
test_df = df.iloc[split_idx:].copy()

print("\n=== train / test 크기 ===")
print("train:", train_df.shape)
print("test :", test_df.shape)

print("\n=== train 기간 ===")
print(train_df["timestamp"].min(), "to", train_df["timestamp"].max())

print("\n=== test 기간 ===")
print(test_df["timestamp"].min(), "to", test_df["timestamp"].max())

# ---------------------------------------------------------
# 11. 평가 함수
# ---------------------------------------------------------
def print_forecast_metrics(y_true, y_pred, model_name="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred) ** 0.5
    try:
        r2 = r2_score(y_true, y_pred)
    except:
        r2 = np.nan

    print(f"\n=== {model_name} 성능 ===")
    print("MAE :", round(mae, 4))
    print("RMSE:", round(rmse, 4))
    print("R²  :", round(r2, 4))

# ---------------------------------------------------------
# 12. Baseline 예측
#    - 바로 이전 값을 그대로 쓰는 naive forecast
# ---------------------------------------------------------
baseline_pred = test_df["target_value"].shift(1)
baseline_pred.iloc[0] = train_df["target_value"].iloc[-1]

print_forecast_metrics(test_df["target_value"], baseline_pred, "Baseline (Naive Forecast)")

plt.figure(figsize=(14, 5))
plt.plot(train_df["timestamp"], train_df["target_value"], label="Train")
plt.plot(test_df["timestamp"], test_df["target_value"], label="Test Actual")
plt.plot(test_df["timestamp"], baseline_pred, label="Baseline Pred", linestyle="--")
plt.title("Baseline Forecast")
plt.xlabel("Date")
plt.ylabel("Target Value")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 13. ARIMA
# ---------------------------------------------------------
arima_model = ARIMA(train_df["target_value"], order=(2, 1, 2))
arima_result = arima_model.fit()

arima_forecast = arima_result.forecast(steps=len(test_df))

print_forecast_metrics(test_df["target_value"], arima_forecast, "ARIMA(2,1,2)")

plt.figure(figsize=(14, 5))
plt.plot(train_df["timestamp"], train_df["target_value"], label="Train")
plt.plot(test_df["timestamp"], test_df["target_value"], label="Test Actual")
plt.plot(test_df["timestamp"], arima_forecast.values, label="ARIMA Forecast", linestyle="--")
plt.title("ARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Target Value")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 14. SARIMA
#    - 계절성까지 반영
# ---------------------------------------------------------
sarima_model = SARIMAX(
    train_df["target_value"],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 30),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarima_result = sarima_model.fit(disp=False)

sarima_forecast = sarima_result.forecast(steps=len(test_df))

print_forecast_metrics(test_df["target_value"], sarima_forecast, "SARIMA(1,1,1)(1,1,1,30)")

plt.figure(figsize=(14, 5))
plt.plot(train_df["timestamp"], train_df["target_value"], label="Train")
plt.plot(test_df["timestamp"], test_df["target_value"], label="Test Actual")
plt.plot(test_df["timestamp"], sarima_forecast.values, label="SARIMA Forecast", linestyle="--")
plt.title("SARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Target Value")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 15. ML용 feature engineering
#    - lag, rolling feature 만들기
# ---------------------------------------------------------
df_ml = df.copy()

# lag features
df_ml["lag_1"] = df_ml["target_value"].shift(1)
df_ml["lag_7"] = df_ml["target_value"].shift(7)
df_ml["lag_14"] = df_ml["target_value"].shift(14)

# rolling features
df_ml["rolling_mean_3"] = df_ml["target_value"].shift(1).rolling(3).mean()
df_ml["rolling_mean_7"] = df_ml["target_value"].shift(1).rolling(7).mean()
df_ml["rolling_std_7"] = df_ml["target_value"].shift(1).rolling(7).std()

# 날짜 기반 feature
df_ml["day_of_week"] = df_ml["timestamp"].dt.dayofweek
df_ml["day_of_month"] = df_ml["timestamp"].dt.day

print("\n=== ML feature 생성 후 상위 10행 ===")
display(df_ml.head(10))

# ---------------------------------------------------------
# 16. ML용 학습 데이터 준비
# ---------------------------------------------------------
feature_cols = [
    "temperature", "pressure", "rpm",
    "lag_1", "lag_7", "lag_14",
    "rolling_mean_3", "rolling_mean_7", "rolling_std_7",
    "day_of_week", "day_of_month"
]

target_col = "target_value"

ml_train_df = df_ml.iloc[:split_idx].copy()
ml_test_df = df_ml.iloc[split_idx:].copy()

X_train_ml = ml_train_df[feature_cols]
y_train_ml = ml_train_df[target_col]

X_test_ml = ml_test_df[feature_cols]
y_test_ml = ml_test_df[target_col]

# 결측치는 lag/rolling 때문에 앞부분에서 생김
imputer = SimpleImputer(strategy="mean")
X_train_ml = pd.DataFrame(imputer.fit_transform(X_train_ml), columns=feature_cols, index=X_train_ml.index)
X_test_ml = pd.DataFrame(imputer.transform(X_test_ml), columns=feature_cols, index=X_test_ml.index)

print("\n=== ML 학습 데이터 크기 ===")
print("X_train_ml:", X_train_ml.shape)
print("X_test_ml :", X_test_ml.shape)

# ---------------------------------------------------------
# 17. ML 회귀 1: Linear Regression
# ---------------------------------------------------------
linear_model = LinearRegression()
linear_model.fit(X_train_ml, y_train_ml)

linear_pred = linear_model.predict(X_test_ml)

print_forecast_metrics(y_test_ml, linear_pred, "ML - Linear Regression")

# ---------------------------------------------------------
# 18. ML 회귀 2: Random Forest Regressor
# ---------------------------------------------------------
rf_model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)
rf_model.fit(X_train_ml, y_train_ml)

rf_pred = rf_model.predict(X_test_ml)

print_forecast_metrics(y_test_ml, rf_pred, "ML - Random Forest Regressor")

# ---------------------------------------------------------
# 19. ML 예측 결과 시각화
# ---------------------------------------------------------
plt.figure(figsize=(14, 5))
plt.plot(train_df["timestamp"], train_df["target_value"], label="Train")
plt.plot(test_df["timestamp"], y_test_ml.values, label="Test Actual")
plt.plot(test_df["timestamp"], linear_pred, label="ML Linear Pred", linestyle="--")
plt.plot(test_df["timestamp"], rf_pred, label="ML RF Pred", linestyle="--")
plt.title("ML-based Time Series Forecast")
plt.xlabel("Date")
plt.ylabel("Target Value")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 20. 모델 성능 비교표
# ---------------------------------------------------------
compare_df = pd.DataFrame({
    "model": [
        "Baseline (Naive)",
        "ARIMA",
        "SARIMA",
        "ML Linear Regression",
        "ML Random Forest"
    ],
    "MAE": [
        mean_absolute_error(test_df["target_value"], baseline_pred),
        mean_absolute_error(test_df["target_value"], arima_forecast),
        mean_absolute_error(test_df["target_value"], sarima_forecast),
        mean_absolute_error(y_test_ml, linear_pred),
        mean_absolute_error(y_test_ml, rf_pred)
    ],
    "RMSE": [
        mean_squared_error(test_df["target_value"], baseline_pred) ** 0.5,
        mean_squared_error(test_df["target_value"], arima_forecast) ** 0.5,
        mean_squared_error(test_df["target_value"], sarima_forecast) ** 0.5,
        mean_squared_error(y_test_ml, linear_pred) ** 0.5,
        mean_squared_error(y_test_ml, rf_pred) ** 0.5
    ],
    "R2": [
        r2_score(test_df["target_value"], baseline_pred),
        r2_score(test_df["target_value"], arima_forecast),
        r2_score(test_df["target_value"], sarima_forecast),
        r2_score(y_test_ml, linear_pred),
        r2_score(y_test_ml, rf_pred)
    ]
})

print("\n=== 모델 성능 비교표 ===")
display(compare_df.sort_values("RMSE"))

# ---------------------------------------------------------
# 21. Random Forest 중요 변수 확인
# ---------------------------------------------------------
rf_importance_df = pd.DataFrame({
    "feature": feature_cols,
    "importance": rf_model.feature_importances_
}).sort_values("importance", ascending=False)

print("\n=== ML Random Forest 중요 변수 ===")
display(rf_importance_df)

plt.figure(figsize=(8, 5))
plt.barh(rf_importance_df["feature"], rf_importance_df["importance"])
plt.gca().invert_yaxis()
plt.title("Feature Importance - ML Random Forest")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()

# ---------------------------------------------------------
# 22. 이상징후 탐색
#    - rolling std 기반 단순 탐지 예시
# ---------------------------------------------------------
threshold = df["rolling_std_7"].mean() + 2 * df["rolling_std_7"].std()
df["anomaly_flag"] = (df["rolling_std_7"] > threshold).astype(int)

print("\n=== 이상징후 개수 ===")
print(df["anomaly_flag"].value_counts())

plt.figure(figsize=(14, 5))
plt.plot(df["timestamp"], df["target_value"], label="target_value")
plt.scatter(
    df.loc[df["anomaly_flag"] == 1, "timestamp"],
    df.loc[df["anomaly_flag"] == 1, "target_value"],
    label="Anomaly Candidate",
    marker="o"
)
plt.title("Simple Anomaly Detection using Rolling Std")
plt.xlabel("Date")
plt.ylabel("Target Value")
plt.legend()
plt.show()

# ---------------------------------------------------------
# 23. 이상 구간 확인
# ---------------------------------------------------------
anomaly_df = df.loc[df["anomaly_flag"] == 1, ["timestamp", "target_value", "rolling_std_7"]].copy()

print("\n=== 이상징후 후보 구간 ===")
display(anomaly_df.head(20))

# ---------------------------------------------------------
# 24. 예측 실패 큰 구간 확인
# ---------------------------------------------------------
error_df = pd.DataFrame({
    "timestamp": test_df["timestamp"].values,
    "actual": test_df["target_value"].values,
    "baseline_pred": baseline_pred.values,
    "arima_pred": arima_forecast.values,
    "sarima_pred": sarima_forecast.values,
    "ml_linear_pred": linear_pred,
    "ml_rf_pred": rf_pred
})

error_df["abs_error_rf"] = (error_df["actual"] - error_df["ml_rf_pred"]).abs()
error_df["abs_error_sarima"] = (error_df["actual"] - error_df["sarima_pred"]).abs()

print("\n=== ML Random Forest 큰 오차 상위 10개 ===")
display(error_df.sort_values("abs_error_rf", ascending=False).head(10))

print("\n=== SARIMA 큰 오차 상위 10개 ===")
display(error_df.sort_values("abs_error_sarima", ascending=False).head(10))

# ---------------------------------------------------------
# 25. 최종 워크숍 질문
# ---------------------------------------------------------
print("\n=== 워크숍 질문 ===")
print("1. Baseline, ARIMA, SARIMA, ML 모델 중 어떤 모델이 가장 좋았는가?")
print("2. RMSE 기준으로 가장 좋은 모델은 무엇인가?")
print("3. 시계열에서는 왜 train/test를 랜덤 분할하면 안 되는가?")
print("4. lag, rolling feature는 왜 필요한가?")
print("5. 이상징후 후보 구간은 어떤 특징을 보이는가?")

# ---------------------------------------------------------
# 26. 6주차 연결 질문
# ---------------------------------------------------------
print("\n=== 다음 시간 생각해볼 질문 ===")
print("- 여러 모델을 조합하면 성능이 더 좋아질까?")
print("- AutoML은 어디까지 자동으로 해줄까?")
print("- 시계열 문제에서도 앙상블이 도움이 될까?")

# ---------------------------------------------------------
# 27. 최종 정리
# ---------------------------------------------------------
print("\n=== 오늘 실습 정리 ===")
print("1. 시계열 데이터 구조를 확인했다.")
print("2. 추세, 계절성, 정상성을 점검했다.")
print("3. ARIMA와 SARIMA를 적용했다.")
print("4. lag, rolling feature를 만든 뒤 ML 회귀 모델도 적용했다.")
print("5. 모델 성능을 비교했다.")
print("6. rolling std를 이용해 간단한 이상징후 후보도 탐색했다.")
```
