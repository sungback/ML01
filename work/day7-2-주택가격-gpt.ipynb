```python
# scikit-learn 의 캘리포니아 주택 가격에 대해 분석, 시각화, 전처리, 머신러닝, 튜닝까지 소스 만들어줘.
```

```python
# ============================================================
# 0. 라이브러리 불러오기
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import platform
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.inspection import permutation_importance

import joblib


# ============================================================
# 1. 기본 설정
# ============================================================

SEED = 42
np.random.seed(SEED)

sns.set_theme(style="whitegrid")

# 한글 폰트 설정
system_name = platform.system()

if system_name == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
elif system_name == "Darwin":   # macOS
    plt.rcParams["font.family"] = "AppleGothic"
else:
    plt.rcParams["font.family"] = "NanumGothic"

plt.rcParams["axes.unicode_minus"] = False


# ============================================================
# 2. 데이터 불러오기
# ============================================================

housing = fetch_california_housing(as_frame=True)

df = housing.frame.copy()

# target 이름을 이해하기 쉽게 변경
df = df.rename(columns={"MedHouseVal": "Price"})

print("데이터 크기:", df.shape)
print()
print("컬럼 목록:")
print(df.columns.tolist())
# ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude', 'Price']
# [ 중위소득, 주택 연령, 평균 방 개수, 평균 침실 개수, 인구수, 평균 가구원 수, 위도, 경도, 주택 가격 (Target) ]


display(df.head())


# ============================================================
# 3. 데이터 기본 정보 확인
# ============================================================

print("\n[데이터 정보]")
print(df.info())

print("\n[기초 통계량]")
display(df.describe().T)

print("\n[결측치 개수]")
display(df.isna().sum())

print("\n[타깃 변수 설명]")
print("Price는 캘리포니아 지역의 주택 가격 중앙값입니다.")
print("단위는 100,000달러입니다.")
print("예: Price = 2.5 → 약 250,000달러")


# ============================================================
# 4. 타깃 변수 분포 시각화
# ============================================================

plt.figure(figsize=(8, 5))
sns.histplot(df["Price"], bins=40, kde=True)
plt.title("주택 가격 분포")
plt.xlabel("주택 가격 중앙값 단위: $100,000")
plt.ylabel("빈도")
plt.tight_layout()
plt.show()

print("Price 최댓값:", df["Price"].max())
print("주의: 이 데이터셋의 Price는 상한값 근처에서 잘린 값이 존재합니다.")


# ============================================================
# 5. 전체 변수 분포 확인
# ============================================================

df.hist(figsize=(14, 10), bins=30)
plt.suptitle("전체 변수 분포", fontsize=16)
plt.tight_layout()
plt.show()


# ============================================================
# 6. 상관관계 분석
# ============================================================

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True
)
plt.title("변수 간 상관관계")
plt.tight_layout()
plt.show()

print("\n[Price와의 상관계수]")
display(
    corr["Price"]
    .sort_values(ascending=False)
    .to_frame("corr_with_price")
)


# ============================================================
# 7. 주요 변수와 가격의 관계 시각화
# ============================================================

important_features = ["MedInc", "AveRooms", "HouseAge", "AveOccup"]

plt.figure(figsize=(14, 10))

for i, col in enumerate(important_features, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(
        data=df.sample(3000, random_state=SEED),
        x=col,
        y="Price",
        alpha=0.4
    )
    plt.title(f"{col} vs Price")

plt.tight_layout()
plt.show()


# ============================================================
# 8. 위도/경도 기반 지역별 가격 시각화
# ============================================================

sample_df = df.sample(5000, random_state=SEED)

plt.figure(figsize=(9, 7))
scatter = plt.scatter(
    sample_df["Longitude"],
    sample_df["Latitude"],
    c=sample_df["Price"],
    s=sample_df["Population"] / 100,
    alpha=0.4,
    cmap="viridis"
)

plt.colorbar(scatter, label="Price")
plt.title("캘리포니아 지역별 주택 가격 분포")
plt.xlabel("경도")
plt.ylabel("위도")
plt.tight_layout()
plt.show()


# ============================================================
# 9. 간단한 파생변수 생성
# ============================================================

df_model = df.copy()

eps = 1e-9

df_model["BedroomsRatio"] = df_model["AveBedrms"] / (df_model["AveRooms"] + eps)
df_model["RoomsPerBedroom"] = df_model["AveRooms"] / (df_model["AveBedrms"] + eps)
df_model["Income_x_HouseAge"] = df_model["MedInc"] * df_model["HouseAge"]

print("파생변수 추가 후 데이터 크기:", df_model.shape)

display(df_model.head())


# ============================================================
# 10. 입력 변수 X, 정답 y 분리
# ============================================================

X = df_model.drop("Price", axis=1)
y = df_model["Price"]

print("X 크기:", X.shape)
print("y 크기:", y.shape)


# ============================================================
# 11. 학습용 / 테스트용 데이터 분리
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=SEED
)

print("학습 데이터:", X_train.shape)
print("테스트 데이터:", X_test.shape)


# ============================================================
# 12. 전처리 파이프라인 구성
# ============================================================

# 이 데이터셋은 전부 수치형 변수입니다.
# 결측치가 없더라도 실무형 코드에서는 SimpleImputer를 넣어두면 안정적입니다.

preprocess = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])


# ============================================================
# 13. 여러 회귀 모델 비교
# ============================================================

models = {
    "Dummy": DummyRegressor(strategy="mean"),
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "RandomForest": RandomForestRegressor(
        n_estimators=200,
        random_state=SEED,
        n_jobs=-1
    ),
    "HistGradientBoosting": HistGradientBoostingRegressor(
        max_iter=300,
        learning_rate=0.05,
        random_state=SEED
    )
}

results = []
trained_models = {}

for name, model in models.items():
    print(f"\n모델 학습 중: {name}")

    pipe = Pipeline([
        ("preprocess", preprocess),
        ("model", model)
    ])

    pipe.fit(X_train, y_train)

    pred = pipe.predict(X_test)

    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2 = r2_score(y_test, pred)

    results.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    })

    trained_models[name] = pipe


results_df = pd.DataFrame(results).sort_values("RMSE")

print("\n[모델 성능 비교]")
display(results_df)


# ============================================================
# 14. 모델 성능 비교 시각화
# ============================================================

plt.figure(figsize=(9, 5))
sns.barplot(
    data=results_df,
    x="RMSE",
    y="Model"
)
plt.title("모델별 RMSE 비교 낮을수록 좋음")
plt.xlabel("RMSE")
plt.ylabel("Model")
plt.tight_layout()
plt.show()

plt.figure(figsize=(9, 5))
sns.barplot(
    data=results_df.sort_values("R2", ascending=False),
    x="R2",
    y="Model"
)
plt.title("모델별 R2 비교 높을수록 좋음")
plt.xlabel("R2")
plt.ylabel("Model")
plt.tight_layout()
plt.show()


# ============================================================
# 15. 하이퍼파라미터 튜닝
# ============================================================

# HistGradientBoostingRegressor는 비교적 빠르고 성능이 좋아서 튜닝 대상으로 사용합니다.

tuning_pipe = Pipeline([
    ("preprocess", preprocess),
    ("model", HistGradientBoostingRegressor(random_state=SEED))
])

param_dist = {
    "model__max_iter": [200, 300, 500],
    "model__learning_rate": [0.03, 0.05, 0.08, 0.1],
    "model__max_leaf_nodes": [15, 31, 63],
    "model__min_samples_leaf": [10, 20, 30, 50],
    "model__l2_regularization": [0.0, 0.01, 0.1, 1.0]
}

search = RandomizedSearchCV(
    estimator=tuning_pipe,
    param_distributions=param_dist,
    n_iter=20,
    scoring="neg_root_mean_squared_error",
    cv=3,
    random_state=SEED,
    n_jobs=-1,
    verbose=1
)

search.fit(X_train, y_train)

print("\n[튜닝 결과]")
print("Best Params:")
print(search.best_params_)
print()
print("Best CV RMSE:", -search.best_score_)


# ============================================================
# 16. 튜닝 모델 테스트 성능 확인
# ============================================================

best_model = search.best_estimator_

best_pred = best_model.predict(X_test)

best_mae = mean_absolute_error(y_test, best_pred)
best_rmse = np.sqrt(mean_squared_error(y_test, best_pred))
best_r2 = r2_score(y_test, best_pred)

print("\n[최종 튜닝 모델 테스트 성능]")
print("MAE :", round(best_mae, 4))
print("RMSE:", round(best_rmse, 4))
print("R2  :", round(best_r2, 4))

final_result = pd.DataFrame([{
    "Model": "Tuned HistGradientBoosting",
    "MAE": best_mae,
    "RMSE": best_rmse,
    "R2": best_r2
}])

results_final_df = pd.concat([results_df, final_result], ignore_index=True)
results_final_df = results_final_df.sort_values("RMSE")

display(results_final_df)


# ============================================================
# 17. 실제값 vs 예측값 시각화
# ============================================================

plt.figure(figsize=(7, 7))
sns.scatterplot(
    x=y_test,
    y=best_pred,
    alpha=0.4
)

min_value = min(y_test.min(), best_pred.min())
max_value = max(y_test.max(), best_pred.max())

plt.plot(
    [min_value, max_value],
    [min_value, max_value],
    linestyle="--",
    color="red"
)

plt.title("실제값 vs 예측값")
plt.xlabel("실제 주택 가격")
plt.ylabel("예측 주택 가격")
plt.tight_layout()
plt.show()


# ============================================================
# 18. 잔차 분석
# ============================================================

residuals = y_test - best_pred

plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=40, kde=True)
plt.title("잔차 분포 실제값 - 예측값")
plt.xlabel("잔차")
plt.ylabel("빈도")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=best_pred,
    y=residuals,
    alpha=0.4
)
plt.axhline(0, linestyle="--", color="red")
plt.title("예측값 대비 잔차")
plt.xlabel("예측값")
plt.ylabel("잔차")
plt.tight_layout()
plt.show()


# ============================================================
# 19. 변수 중요도 분석
# ============================================================

# permutation_importance는 모델 종류와 관계없이 사용할 수 있는 중요도 분석 방법입니다.
# 계산 시간이 걸릴 수 있으므로 테스트 데이터 일부만 사용합니다.

X_test_sample = X_test.sample(3000, random_state=SEED)
y_test_sample = y_test.loc[X_test_sample.index]

importance = permutation_importance(
    best_model,
    X_test_sample,
    y_test_sample,
    n_repeats=5,
    random_state=SEED,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)

importance_df = pd.DataFrame({
    "Feature": X.columns,
    "Importance": importance.importances_mean
}).sort_values("Importance", ascending=False)

print("\n[변수 중요도]")
display(importance_df)

plt.figure(figsize=(9, 6))
sns.barplot(
    data=importance_df,
    x="Importance",
    y="Feature"
)
plt.title("변수 중요도 Permutation Importance")
plt.xlabel("중요도")
plt.ylabel("변수")
plt.tight_layout()
plt.show()


# ============================================================
# 20. 예측 결과 샘플 확인
# ============================================================

pred_result = X_test.copy()
pred_result["Actual"] = y_test
pred_result["Predicted"] = best_pred
pred_result["Error"] = pred_result["Actual"] - pred_result["Predicted"]
pred_result["AbsError"] = pred_result["Error"].abs()

display(
    pred_result[["Actual", "Predicted", "Error", "AbsError"]]
    .sort_values("AbsError", ascending=False)
    .head(10)
)

display(
    pred_result[["Actual", "Predicted", "Error", "AbsError"]]
    .sample(10, random_state=SEED)
)


# ============================================================
# 21. 최종 모델 저장
# ============================================================

joblib.dump(best_model, "california_housing_best_model.pkl")

print("\n모델 저장 완료: california_housing_best_model.pkl")


# ============================================================
# 22. 저장한 모델 다시 불러와서 예측 테스트
# ============================================================

loaded_model = joblib.load("california_housing_best_model.pkl")

sample_input = X_test.head(5)
sample_pred = loaded_model.predict(sample_input)

print("\n[저장된 모델 예측 테스트]")
display(pd.DataFrame({
    "Actual": y_test.head(5).values,
    "Predicted": sample_pred
}))
```
