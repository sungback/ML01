```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# 1. 데이터 로드 및 분할 (학습용 80%, 테스트용 20%)
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=42
)

# 2. 비교할 모델 설정
models = {
    "Linear": SVC(kernel='linear', C=1.0),
    "Polynomial": SVC(kernel='poly', degree=3, C=1.0, gamma='auto'),
    "RBF": SVC(kernel='rbf', C=1.0, gamma='auto')
}

# 3. 결과 출력
print(f"{'Kernel Type':<20} | {'Accuracy':<10}")
print("-" * 35)

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name:<20} | {acc:.4f}")

# 상세 리포트 (RBF 기준)
print("\n[RBF 모델 상세 리포트]")
print(classification_report(y_test, models["RBF"].predict(X_test), target_names=iris.target_names))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# 1. 테스트할 파라미터 조합 설정
param_grid = {
    'C': [0.1, 1, 10, 100],
    'degree': [2, 3, 4],
    'gamma': ['scale', 'auto', 0.1, 1],
    'kernel': ['poly'] # poly 커널에 대해서만 테스트
}

# 2. 그리드 서치 객체 생성 (5-겹 교차 검증)
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)

# 3. 학습 (가장 좋은 조합을 찾음)
grid.fit(X_train, y_train)

# 4. 최적의 파라미터 출력
print(f"최적 파라미터: {grid.best_params_}")
print(f"최고 정확도: {grid.best_score_:.4f}")

# 5. 최적의 모델로 바로 예측 가능
best_model = grid.best_estimator_
# 최적 파라미터: {'C': 0.1, 'degree': 2, 'gamma': 'auto', 'kernel': 'poly'}
# 최고 정확도: 0.9583
```
