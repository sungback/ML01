```python
import joblib  # 모델 저장 및 로드를 위한 라이브러리 (pickle보다 대용량 처리에 최적화)
from sklearn import datasets  # 붓꽃(Iris) 데이터셋을 가져오기 위함
from sklearn.svm import SVC   # 서포트 벡터 머신(SVM) 분류 모델 사용


# 1. 데이터 준비
iris = datasets.load_iris()
X = iris.data    # 꽃받침/꽃잎의 수치 데이터 (입력값)
y = iris.target  # 꽃의 종류 (정답값: 0, 1, 2)

# 2. 모델 생성 및 설정
# probability=True: "이 꽃은 90% 확률로 세토사야"처럼 확률 정보를 계산하게 함
model = SVC(
    kernel='poly',   # 곡선 형태의 경계선을 사용함
    degree=3,        # 3차 함수 곡선으로 복잡도를 조절함
    C=1.0,           # 오차를 얼마나 허용할지 결정 (1.0은 표준적인 규제)
    gamma='auto',    # 데이터 영향력의 범위를 자동으로 설정
    probability=True,
    random_state=42  # 실행할 때마다 결과가 달라지지 않도록 고정
)

# 3. 모델 학습 (AI에게 데이터를 공부시킴)
model.fit(X, y)

# 4. 학습된 모델을 파일로 저장
# 확장자는 관습적으로 .pkl 또는 .joblib을 사용합니다.
model_filename = 'svm_model.pkl'
joblib.dump(model, model_filename)

print(f"✅ 학습 완료! '{model_filename}' 파일이 생성되었습니다.")
```
